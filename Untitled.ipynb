```python
import pandas as pd
import numpy as np
import yfinance as yf

class FinancialData:
    def __init__(self, ticker):
        self.ticker = ticker
        self.data = None
    
    def download_data(self):
        self.data = yf.download(self.ticker, period="10y", interval="1d")
        self.data['symbol'] = self.ticker
        
    def wilder_smoothing(self, data, periods):
        wilder = np.empty_like(data)
        wilder[:] = np.nan
        if len(data) < periods or np.isnan(data.iloc[:periods]).all():
            return wilder
        wilder[periods-1] = np.nanmean(data.iloc[:periods])
        for i in range(periods, len(data)):
            if np.isnan(wilder[i-1]) or np.isnan(data.iloc[i]):
                continue
            wilder[i] = (wilder[i-1] * (periods - 1) + data.iloc[i]) / periods
        return wilder
    
    def calculate_indicators(self):
        # Bollinger Bands
        self.data['middle_band'] = self.data['Close'].rolling(window=20).mean()
        self.data['std_dev'] = self.data['Close'].rolling(window=20).std()
        self.data['upper_band'] = self.data['middle_band'] + (self.data['std_dev'] * 2)
        self.data['lower_band'] = self.data['middle_band'] - (self.data['std_dev'] * 2)

        # MACD
        self.data['ShortEMA'] = self.data['Close'].ewm(span=12, adjust=False).mean()
        self.data['LongEMA'] = self.data['Close'].ewm(span=26, adjust=False).mean()
        self.data['MACD'] = self.data['ShortEMA'] - self.data['LongEMA']
        self.data['signal'] = self.data['MACD'].ewm(span=9, adjust=False).mean()

        # RSI
        delta = self.data['Close'].diff()
        up = delta.clip(lower=0)
        down = -1 * delta.clip(upper=0)
        ema_up = up.ewm(com=13, adjust=False).mean()
        ema_down = down.ewm(com=13, adjust=False).mean()
        rs = ema_up / ema_down
        self.data['RSI'] = 100 - (100 / (1 + rs))

        # ATR and ADX are used to determine signal strength and a true range, for volatility
        self.data['prev_close'] = self.data.groupby('symbol')['Close'].shift(1)
        self.data['TR'] = np.maximum((self.data['High'] - self.data['Low']), 
                            np.maximum(abs(self.data['High'] - self.data['prev_close']), 
                            abs(self.data['prev_close'] - self.data['Low'])))

        for i in self.data['symbol'].unique():
            TR_data = self.data[self.data.symbol == i].copy()
            self.data.loc[self.data.symbol==i,'ATR_5'] = self.wilder_smoothing(TR_data['TR'], 5)
            self.data.loc[self.data.symbol==i,'ATR_15'] = self.wilder_smoothing(TR_data['TR'], 15)
        self.data['ATR_Ratio'] = self.data['ATR_5'] / self.data['ATR_15']

        # This code for ATR and ADX  is from the internet 
        self.data['prev_high'] = self.data.groupby('symbol')['High'].shift(1)
        self.data['prev_low'] = self.data.groupby('symbol')['Low'].shift(1)

        self.data['+DM'] = np.where(~np.isnan(self.data.prev_high),
                                np.where((self.data['High'] > self.data['prev_high']) & 
                                (((self.data['High'] - self.data['prev_high']) > (self.data['prev_low'] - self.data['Low']))), 
                                self.data['High'] - self.data['prev_high'], 0),np.nan)
        self.data['-DM'] = np.where(~np.isnan(self.data.prev_low),
                                np.where((self.data['prev_low'] > self.data['Low']) & 
                                (((self.data['prev_low'] - self.data['Low']) > (self.data['High'] - self.data['prev_high']))), 
                                self.data['prev_low'] - self.data['Low'], 0),np.nan)

        for i in self.data['symbol'].unique():
            ADX_data = self.data[self.data.symbol == i].copy()
            self.data.loc[self.data.symbol==i,'+DM_15'] = self.wilder_smoothing(ADX_data['+DM'], 15)
            self.data.loc[self.data.symbol==i,'-DM_15'] = self.wilder_smoothing(ADX_data['-DM'], 15)
        self.data['+DI_15'] = (self.data['+DM_15']/self.data['ATR_15'])*100
        self.data['-DI_15'] = (self.data['-DM_15']/self.data['ATR_15'])*100
        self.data['DX_15'] = (np.round(abs(self.data['+DI_15'] - self.data['-DI_15'])/(self.data['+DI_15'] + self.data['-DI_15']) * 100))

        for i in self.data['symbol'].unique():
            ADX_data = self.data[self.data.symbol == i].copy()
            self.data.loc[self.data.symbol==i,'ADX_15'] = self.wilder_smoothing(ADX_data['DX_15'], 15)

        # Determine if the stock has momentum and volatility (strength of the signal)
        self.data['adx_signal'] = self.data.apply(lambda x : 1 if (x['ADX_15'] >= 25) else 0, axis = 1)
        self.data['atr_signal'] = self.data.apply(lambda x : 1 if (x['ATR_Ratio'] >= 1) else 0, axis = 1)

    def weekly_resample(self):
        # Resampling data on a weekly basis
        self.weekly_data = self.data.select_dtypes(include=[np.number]).resample('W').mean()
```

```python
class FeatureEngineering:
    def __init__(self, data):
        self.data = data
    
    def add_features(self):
        # Choosing threshold to start with arbitrarily, can optimize later
        self.low_atr_threshold = self.data['ATR_15'].quantile(0.25)
        self.data['is_consolidating'] = (self.data['ADX_15'] < 20) & (self.data['ATR_15'] < self.low_atr_threshold)
        self.data['is_consolidating'] = self.data['is_consolidating'].astype(bool)
        # Calculates duration of stability for scoring, ffill for carrying it forward
        self.data['consolidation_start_date'] = self.data.index.where(self.data['is_consolidating'] & (~self.data['is_consolidating'].shift(1).fillna(False)))
        self.data['consolidation_start_date'] = self.data['consolidation_start_date'].ffill()
        self.data['consolidation_duration'] = (self.data.index - self.data['consolidation_start_date']).dt.days # Extracts number of days from Timedelta
        # Gets range from bollinger bands
        self.data['range'] = self.data['upper_band'] - self.data['lower_band']
        self.data['avg_range'] = self.data['range'].rolling(window=10).mean()
        # Avg Volume used for scoring
        self.data['avg_volume'] = self.data['Volume'].rolling(window=10).mean()
        
    def calculate_scores(self):
        # Calculates bollinger position score and Range Score (similar)
        self.data['bollinger_position_score'] = (self.data['Close'] - self.data['lower_band']) / (self.data['upper_band'] - self.data['lower_band'])
        self.data['range_score'] = 1 - (self.data['range'] / self.data['avg_range'])
        # Volume Score
        self.data['volume_score'] = 1 - (self.data['Volume'] / (self.data['avg_volume'] + 1e-10))
        # Consolidation Score + Consolidation Duration Score
        self.data['consolidation_score'] = (self.data['upper_band'] - self.data['lower_band']) / self.data['ATR_15']
        self.data['duration_score'] = (self.data['consolidation_duration'] - self.data['consolidation_duration'].min()) / (self.data['consolidation_duration'].max() - self.data['consolidation_duration'].min())
        self.data['duration_score'] = self.data['duration_score'].fillna(0)
        # 2 Trend scores will account for the strength of the signals
        self.data['smoothed_trend_score'] = self.data['MACD'] - self.data['signal']
        self.data['smoothed_trend_score1'] = (self.data['smoothed_trend_score'] - self.data['smoothed_trend_score'].rolling(window=10).mean()) / self.data['smoothed_trend_score'].rolling(window=10).std()
        self.data['trend_score'] = 0
        # Trend Score 2 - Uptrend indicators 
        self.data.loc[(self.data['ShortEMA'] > self.data['LongEMA']) & (self.data['MACD'] > self.data['signal']), 'trend_score'] += 1
        self.data.loc[self.data['RSI'] > 60, 'trend_score'] += 1  # Consider lowering the threshold for more sensitivity
        self.data.loc[(self.data['ADX_15'] > 25) & (self.data['MACD'] > self.data['signal']), 'trend_score'] += 1  # Strong trend
        self.data.loc[(self.data['ATR_Ratio'] > 1) & (self.data['MACD'] > self.data['signal']), 'trend_score'] += 1  # Higher current volatility vs. past
        #  Trend Score 2 - Downtrend indicators
        self.data.loc[(self.data['ShortEMA'] < self.data['LongEMA']) & (self.data['MACD'] < self.data['signal']), 'trend_score'] -= 1
        self.data.loc[self.data['RSI'] < 40, 'trend_score'] -= 1  # Consider raising the threshold for more sensitivity
        self.data.loc[(self.data['ADX_15'] > 25)  & (self.data['MACD'] < self.data['signal']), 'trend_score'] -= 1  # Strong trend
        self.data.loc[(self.data['ATR_Ratio'] < 1) & (self.data['MACD'] < self.data['signal']), 'trend_score'] -= 1  # Lower current volatility vs. past
        # Normalize the trend score to keep it bounded between -1 and 1 for classification
        self.data['trend_score_normalized'] = self.data['trend_score'] / self.data['trend_score'].abs().max()
        self.data['smoothed_trend_score2'] = self.data['trend_score_normalized'].rolling(window=10).mean()
        # Combined scores
        self.data['combined_score'] = (self.data['range_score'] * 0.30 +
                                   self.data['duration_score'] * 0.13 +
                                   self.data['volume_score'] * 0.2 +
                                  (self.data['ADX_15'] < 20) * 0.185 +  # Use binary signals for low ADX
                                  (self.data['ATR_15'] < self.low_atr_threshold) * 0.185)
        # Calculate a simple moving average of the consolidation score with a window size of 10 days
        self.data['smoothed_combined_score'] = self.data['combined_score'].rolling(window=10).mean()
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from sklearn.impute import SimpleImputer
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

class ModelTester:
    def __init__(self, data):
        self.data = data

    def scale_data(self, features):
        """Scales the data for the given features after imputing missing values."""
        imputer = SimpleImputer(strategy='mean')
        data_imputed = imputer.fit_transform(self.data[features])
        scaler = StandardScaler()
        return scaler.fit_transform(data_imputed)

    def fit_kmeans(self, data_scaled, n_clusters):
        """Fits K-Means clustering and returns inertia."""
        clusterer = KMeans(n_clusters=n_clusters, random_state=10, n_init = 10)
        cluster_labels = clusterer.fit_predict(data_scaled)
        return clusterer.inertia_

    def fit_agglomerative(self, data_scaled, n_clusters):
        """Fits Agglomerative Clustering and returns the cluster labels and silhouette score."""
        clusterer = AgglomerativeClustering(n_clusters=n_clusters)
        cluster_labels = clusterer.fit_predict(data_scaled)
        silhouette_avg = silhouette_score(data_scaled, cluster_labels)
        return cluster_labels, silhouette_avg

    def find_optimal_clusters(self, features):
        """Determines the optimal number of clusters using the silhouette and elbow methods for KMeans and applies Agglomerative Clustering."""
        data_scaled = self.scale_data(features)
        range_n_clusters = range(2, 11)
        elbow = []
        silhouette_scores = []

        for n_clusters in range_n_clusters:
            inertia = self.fit_kmeans(data_scaled, n_clusters)
            elbow.append(inertia)
            _, silhouette = self.fit_agglomerative(data_scaled, n_clusters)
            silhouette_scores.append(silhouette)

        optimal_k_silhouette = np.argmax(silhouette_scores) + 2
        optimal_k_elbow = self.find_elbow(elbow)

        labels_silhouette, silhouette_silhouette = self.fit_agglomerative(data_scaled, optimal_k_silhouette)
        labels_elbow, silhouette_elbow = self.fit_agglomerative(data_scaled, optimal_k_elbow)

        # Extract the inertia for the optimal k determined by the elbow method
        elbow_inertia = elbow[optimal_k_elbow - 2]  # -2 because range starts at 2

        return optimal_k_silhouette, optimal_k_elbow, silhouette_silhouette, silhouette_elbow, labels_silhouette, labels_elbow, elbow_inertia


    def find_elbow(self, elbow):
        """Calculates the elbow point in the inertia plot."""
        y = np.array(elbow)
        n_points = len(y)
        all_coords = np.vstack((range(n_points), y)).T
        first_point = all_coords[0]
        line_vec = all_coords[-1] - all_coords[0]
        line_vec_norm = line_vec / np.sqrt(np.sum(line_vec**2))
        vec_from_first = all_coords - first_point
        scalar_product = np.sum(vec_from_first * np.tile(line_vec_norm, (n_points, 1)), axis=1)
        vec_from_first_parallel = np.outer(scalar_product, line_vec_norm)
        vec_to_line = vec_from_first - vec_from_first_parallel
        dist_to_line = np.sqrt(np.sum(vec_to_line ** 2, axis=1))
        idx_of_elbow = np.argmax(dist_to_line)
        return idx_of_elbow + 2

    def test_features(self):
        """Tests different feature sets and records results, including optimal k and cluster labels."""
        feature_sets = [
            ['ATR_Ratio', 'MACD', 'ADX_15', 'consolidation_score', 'smoothed_trend_score', 'volume_score'],
            ['ATR_Ratio', 'MACD', 'ADX_15', 'consolidation_score', 'smoothed_trend_score']  # No volume_score
            # More combinations can be added
        ]
        results = {}
        for features in feature_sets:
            optimal_k_silhouette, optimal_k_elbow, silhouette_silhouette, silhouette_elbow, labels_silhouette, labels_elbow, elbow_inertia = self.find_optimal_clusters(features)
            results[tuple(features)] = {
                'Optimal K by Silhouette': optimal_k_silhouette,
                'Silhouette by Silhouette': silhouette_silhouette,
                'Labels by Silhouette': labels_silhouette,
                'Optimal K by Elbow': optimal_k_elbow,
                'Silhouette by Elbow': silhouette_elbow,
                'Labels by Elbow': labels_elbow,
                'Elbow Inertia': elbow_inertia }
        return results
    
    def plot_clusters(self, data, labels, title):
        """Plots the data with clusters visualized."""
        fig, ax = plt.subplots(figsize=(14, 7))
        # Define color scheme for clusters
        colors = ['red', 'green', 'blue', 'purple', 'orange', 'yellow']  # Expand as needed
        unique_labels = np.unique(labels)

        # Plot each cluster
        for label, color in zip(unique_labels, colors[:len(unique_labels)]):
            clustered_data = data[data['Cluster'] == label]
            ax.plot(clustered_data.index, clustered_data['Close'], label=f'Cluster {label}', color=color, alpha=0.5)

        # Highlight the background according to cluster
        for i, color in enumerate(colors[:len(unique_labels)]):
            ax.fill_between(data.index, data['Close'].min(), data['Close'].max(),
                            where=(labels == i),
                            facecolor=color, alpha=0.2, label=f'Cluster {i}')

        ax.set_title(title)
        ax.set_xlabel('Date')
        ax.set_ylabel('Close Price')
        ax.xaxis.set_major_locator(mdates.MonthLocator())
        ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
        ax.set_ylim(data['Close'].min() * 0.95, data['Close'].max() * 1.05)
        ax.legend()
        plt.tight_layout()
        plt.show()
        
    def choose_best_clustering(results):
        best_labels = None
        best_score = -1

        # Evaluate based on silhouette score primarily
        for features, result in results.items():
            silhouette_silhouette = result['Silhouette by Silhouette']
            silhouette_elbow = result['Silhouette by Elbow']

            # You could also factor in inertia by converting it to a comparable scale or through a scoring function
            if silhouette_silhouette > best_score:
                best_score = silhouette_silhouette
                best_labels = result['Labels by Silhouette']
            elif silhouette_silhouette == best_score:  # Tie case, decide by another metric like inertia
                # Placeholder for inertia comparison, depends on conversion or scoring function
                pass

        return best_labels
```

```python
def normalize(values):
    """ Normalize to [0, 1] scale. """
    min_val = min(values)
    max_val = max(values)
    return [(val - min_val) / (max_val - min_val) if max_val != min_val else 0 for val in values]

def choose_best_clustering(results):
    silhouettes = [result['Silhouette by Silhouette'] for result in results.values()]
    inertias = [result['Elbow Inertia'] for result in results.values()]

    # Normalize both metrics
    normalized_silhouettes = normalize(silhouettes)
    normalized_inertias = normalize([-inertia for inertia in inertias])  # Negative because lower is better

    # Weights can be adjusted
    silhouette_weight = 0.7
    inertia_weight = 0.3

    # Calculate combined scores
    combined_scores = [silhouette_weight * silhouette + inertia_weight * inertia 
                       for silhouette, inertia in zip(normalized_silhouettes, normalized_inertias)]

    # Find the index of the best score
    best_index = combined_scores.index(max(combined_scores))

    # Return the corresponding labels
    key_of_best = list(results.keys())[best_index]
    best_labels = results[key_of_best]['Labels by Silhouette'] if normalized_silhouettes[best_index] > normalized_inertias[best_index] else results[key_of_best]['Labels by Elbow']
    
    return best_labels, key_of_best  

def main():
    tickers = 'SPY'
    # Process daily data
    fd_daily = FinancialData(tickers)
    fd_daily.download_data()
    fd_daily.calculate_indicators()

    fe_daily = FeatureEngineering(fd_daily.data)
    fe_daily.add_features()
    fe_daily.calculate_scores()

    tester_daily = ModelTester(fd_daily.data)
    results_daily = tester_daily.test_features()
    best_labels_daily, best_key_daily = choose_best_clustering(results_daily)
    fd_daily.data['Cluster'] = best_labels_daily

    # Process weekly data
    fd_weekly = FinancialData(tickers)
    fd_weekly.download_data()
    fd_weekly.calculate_indicators()
    fd_weekly.weekly_resample()

    fe_weekly = FeatureEngineering(fd_weekly.weekly_data)
    fe_weekly.add_features()
    fe_weekly.calculate_scores()

    tester_weekly = ModelTester(fd_weekly.weekly_data)
    results_weekly = tester_weekly.test_features()
    best_labels_weekly, best_key_weekly = choose_best_clustering(results_weekly)
    fd_weekly.weekly_data['Cluster'] = best_labels_weekly

    # Visualization of clusters
    tester_daily.plot_clusters(fd_daily.data, fd_daily.data['Cluster'], "Daily Data Clusters")
    tester_weekly.plot_clusters(fd_weekly.weekly_data, fd_weekly.weekly_data['Cluster'], "Weekly Data Clusters")

    # Save results to a text file
    with open('clustering_results.txt', 'w') as file:
        file.write("Daily Data Results:\n")
        file.write(str(results_daily) + "\n")
        file.write(f"Best Performance by {best_key_daily}\n\n")
        file.write("Weekly Aggregated Data Results:\n")
        file.write(str(results_weekly))
        file.write(f"Best Performance by {best_key_weekly}\n")

if __name__ == "__main__":
    main()
```
